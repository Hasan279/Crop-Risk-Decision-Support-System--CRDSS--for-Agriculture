# Week 1 Exploration
Visualizing weather risks across different agricultural regions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn style
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../data/processed/engineered_features.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

## 1. Frost Events
Total frost events per city over the entire time period.

```python
frost_counts = df.groupby('city_name')['frost_event'].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=frost_counts.index, y=frost_counts.values, palette='Blues_r')
plt.title('Total Frost Events per Region (2005-2026)')
plt.ylabel('Number of Days')
plt.xlabel('Region')
plt.show()
```

## 2. Heat Stress Events
Total heat stress events per city (days part of a 3-day consecutive >35�C streak).

```python
heat_counts = df.groupby('city_name')['heat_stress_event'].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=heat_counts.index, y=heat_counts.values, palette='Reds_r')
plt.title('Total Heat Stress Events per Region (2005-2026)')
plt.ylabel('Number of Days')
plt.xlabel('Region')
plt.show()
```

## 3. Heavy Rain Events
Total heavy rain events per city (>50mm over 3 days).

```python
rain_counts = df.groupby('city_name')['heavy_rain_event'].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=rain_counts.index, y=rain_counts.values, palette='Greens_r')
plt.title('Total Heavy Rain Events per Region (2005-2026)')
plt.ylabel('Number of Days')
plt.xlabel('Region')
plt.show()
```
